### Selecting the question columns that we will use that hold a 1-5 rating, and replacing 0's and 6's with NaNs.

```python
questions=['SFORate-Restaurants',
           'SFORate-Stores',
           'SFORate-Signs',
           'SFORate-MovingWalkways',
           'SFORate-ScreenInformation',
           'SFORate-InformationBooths(Lower)',
           'SFORate-InformationBooths(Upper)',
           'SFORate-ParkingFacilities',
           'SFOClean-BoardingAreas',
           'SFOClean-Parking',
           'SFOClean-AirTrain',
           'SFOClean-RentalCarCenter',
           'SFOClean-Restaurants',
           'SFOClean-Restrooms',]
```

```python
questions_1=['SFORate-Restaurants',
             'SFORate-Stores',
             'SFORate-Signs',
             'SFORate-MovingWalkways',
             'SFORate-ScreenInformation',
             'SFORate-InformationBooths(Lower)',
             'SFORate-InformationBooths(Upper)',
             'SFORate-ParkingFacilities']
```

#### Replacing the 0 and 6 values with NAN to make data consistent.

```python
for i in questions:
    customer[i] = customer[i].replace([0,6],[np.nan,np.nan])
customer[questions].head(n=5)
```

```python
for i in questions_1:   
    customer_12[i] = customer_12[i].replace([0,6],[np.nan,np.nan])
    customer_13[i] = customer_13[i].replace([0,6],[np.nan,np.nan])
```
