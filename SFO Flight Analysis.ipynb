## SFO PASSENGERS

With this question we are trying to find a Least Squares Regression between years and the total number of passengers, and see what the trend looks like from 2005-2014. Following that, check whether there is any correlation between the two parameters.

To calculate the total number of passengers from 2005-2014, we have grouped the data by year, which gave us a count of how many customers flew each year from 2005 to 2014. We have shown this data using a scatter plot, with Year on the X-axis and Total Passenger Count on the Y-axis.

Assumption: The Dataset we have is from 2005-2014; however, the total passenger count in 2005 and 2014 is not complete. 2014 is still not over, and recording the flight information began in July 2005.

#### Grouping Data and Scatter Plotting

```python
ByYear = flights.groupby(['Year']).sum()
ByYear = ByYear.ix[:,2:]
np.array(ByYear['Adjusted Passenger Count'])
ByYear = ByYear.reindex (columns = ['Years','Adjusted Passenger Count'])
ByYear.Years = [2005,2006,2007,2008,2009,2010,2011,2012,2013,2014]
ByYear1 = pd.DataFrame(ByYear)
```

```python
ByYear1.plot(kind='scatter',y='Adjusted Passenger Count',x='Years',color='c')
plt.xlabel('Year')
plt.ylabel('Total Passenger Count in 10 Mn');
```

Now, we have put the two parameters in an array to calculate the Least Squares Regreesion, which results in a p-value that indicates whether the data suports the Null Hypothesis or not. r-value and r-squared values help determine the relationship between the independent and dependent variable

```python
count1= np.array(ByYear1['Adjusted Passenger Count'])
count1 = count1.astype(np.float32)
years1=np.array(ByYear1['Years'])
years1 = years1.astype(np.int)
```

#### Calculating Least Squares Regression Returns: Slope, Intercept, p-value, and R squared

```python
a= count1
b= years1

slope, intercept, r_value, p_value, std_err = stats.linregress(b,a)
print ("=============================================")
print("Slope- " + str(slope))
print("Intercept" + str(intercept))
print("r_value- " + str(r_value))
print("p_value: " + str(p_value))
print("R square: " + str(r_value*r_value))
print ("=============================================")
ByYear1.plot(kind='scatter',y='Adjusted Passenger Count',x='Years',color='crimson')
rating= np.linspace(2005,2014,9)             
plt.plot(rating,intercept + (slope*rating), 'b')
plt.axis([2004,2016,1500000,46000000])
plt.xlabel('Year')
plt.ylabel('Total Passenger Count in 10 Mn')
plt.title('Regression plot for total Passengers Over the Years');
```

From the graph and values above, we can conclude that there is a strong correlation between the total number of passengers by year. The blue line represents the regression line, showing us that the number of passengers is consistently increasing every year, and it is expected to continue to grow in the coming years.

Furthermore, we got an r-squared value around 50%, which indicates that the relationship between the independent variable (Year) and dependent variable (Adjusted Passenger Count) is strong. The p-value is also good and in the range of 0-0.05, which shows that it strongly supports the Null Hypothesis.

In this scenario, the Year is an independent variable and Total Passenger Count is dependent. We can see that 2009, the number of passengers flying dropped, because of the recent economic recession. However, SFO still maintained a strong regression, and SFO's total passenger count went back up in 2010.
